```python
import pandas as pd
import pathlib
```

```python
p = pathlib.Path(r'C:\Users\nyzhur\Git\fmcgprep\fmcgprep\fmcg_for_Lenta')
paths = [x for x in p.iterdir() if x.is_file()]


```

```python
full_df = pd.DataFrame()
for path in  paths:
    tmp_df = pd.read_excel(path)
    hh_filter = path.name.split('__')[-1]
    tmp_df['hh_filter'] =[hh_filter]*len(tmp_df)
    full_df =pd.concat([full_df,tmp_df] )
```

```python
full_df
```

```python
full_df.groupby('hh_filter')['month_end'].nunique()
```

```python
metrics = """value	volume_gl	volume_pack	trips_raw	trips	buyers	buyers_raw	population	penetration	frequency	spend_per_buyer	spend_per_trip	volume_per_buyer_g	volume_per_buyer_pack	volume_per_trip_g	volume_per_trip_pack	avg_price_g	avg_price_pack	avg_pack_size""".split('\t')
```

```python
full_df[list(set(full_df.columns)- set(metrics))]
```

```python
pivot =pd.pivot_table(full_df, index=['position_number','position_name','hh_filter', ], columns = ['month_end'],  values = metrics )
```

```python
pivot
```

```python
with pd.ExcelWriter(p.joinpath('pivot.xlsx')) as writer:  
    full_df.to_excel(writer, sheet_name='Data')
    pivot.to_excel(writer, sheet_name='Pivot')
```
